# Coletando os dados na origem [Lista de bancos] | Collecting the data at the source [List of banks]

```python
# Lendo o HTML e selecionando o primeiro arquivo encontrado | Reading the HTML and selecting the first file found
lista_falencia_bancos = pd.read_html('https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/')[0]
```

```python
lista_falencia_bancos.info()
```

```python
# Renomeando as colunas | Renaming columns
lista_falencia_bancos.rename(columns = {"Bank NameBank": "nome_do_banco", 
                                        "CityCity": "cidade",
                                        "StateSt": "estado",
                                        "CertCert": "cert",
                                        "Acquiring InstitutionAI": "adquirente",
                                        "Closing DateClosing": "data_de_fechamento",
                                        "FundFund": "Fund"
                                        }, inplace = True)
```

```python
# Ajustando o tipo da data | Adjusting the date type
lista_falencia_bancos['data_de_fechamento'] = pd.to_datetime(lista_falencia_bancos['data_de_fechamento'])
```

```python
lista_falencia_bancos
```

```python
# Subindo o dado para a nuvem | Uploading data to the cloud
pandas_gbq.to_gbq(lista_falencia_bancos, '[REMOVED FOR SECURITY]', project_id='[REMOVED FOR SECURITY]', if_exists='replace')
```

# Coletando os dados na origem [Assets] | Collecting the data at the source [Assets]

```python
# Lendo o HTML e selecionando o primeiro arquivo encontrado | Reading the HTML and selecting the first file found
lista_falencia_assets = pd.read_html('https://www.fdic.gov/bank/historical/bank/')[0]
```

```python
lista_falencia_assets
```

```python
lista_falencia_assets.info()
```

```python
# Preparando a tabela para o transpose | Preparing the table for transpose
lista_falencia_assets.set_index('Years', inplace = True)
```

```python
# Transpondo os dados para melhor manipulação | Transposing the data for better handling
lista_falencia_assets = lista_falencia_assets.transpose()
```

```python
lista_falencia_assets
```

```python
#Renomeando as colunas para o padrão do banco de dados | Renaming columns to the database default
lista_falencia_assets.rename(columns = {'Bank Failures': 'bank_failures',
                                        'Total Assets (Millions)': 'total_assets_millions', 
                                        }, inplace = True)
```

```python
# Reindexando para transformar o ano em uma coluna dataframe | Reindexing to turn the year into a dataframe column
lista_falencia_assets.reset_index(inplace = True, names = 'year')
```

```python
lista_falencia_assets
```

```python
# Ajustando o tipo de dados | Adjusting the data type
lista_falencia_assets['bank_failures'] = lista_falencia_assets['bank_failures'].astype('int')
lista_falencia_assets['year'] = lista_falencia_assets['year'].astype('int')
```

```python
lista_falencia_assets.info()
```

```python
# Subindo o dado para a nuvem | Uploading data to the cloud
pandas_gbq.to_gbq(lista_falencia_assets, '[REMOVED FOR SECURITY]', project_id='[REMOVED FOR SECURITY]', if_exists='replace')
```
